```python
import firebase_admin
from firebase_admin import credentials, db
import json

# Load the private key
key_data = None
try:
    with open("firebase_key.json", "r", encoding="utf-8") as f:
        key_data = json.load(f)
except UnicodeDecodeError:
    # If UTF-8 fails, try latin-1 as a fallback for reading the bytes as characters
    print("Warning: firebase_key.json could not be decoded with UTF-8. Trying latin-1 encoding.")
    try:
        with open("firebase_key.json", "r", encoding="latin-1") as f:
            content = f.read()
        key_data = json.loads(content) # Attempt to parse the latin-1 decoded string as JSON
    except json.JSONDecodeError as e:
        print(f"Error: firebase_key.json is not a valid JSON file even with latin-1 decoding: {e}")
        # Re-raise the error as we can't proceed without a valid key
        raise
    except Exception as e:
        print(f"An unexpected error occurred while decoding with latin-1: {e}")
        raise
except json.JSONDecodeError as e:
    print(f"Error: firebase_key.json is not a valid JSON file with UTF-8 encoding: {e}")
    # Re-raise the error as we can't proceed without a valid key
    raise
except Exception as e:
    print(f"An unexpected error occurred while decoding with UTF-8: {e}")
    raise

# If key_data was successfully loaded as a dictionary, use it to create credentials
if key_data:
    cred = credentials.Certificate(key_data)

    # Initialize the app with your database URL
    firebase_admin.initialize_app(cred, {
     "databaseURL": "https://rosal-demesa-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"
    })
    print("Firebase connected successfully!")
else:
    print("Firebase initialization failed: Could not load service account key.")
```

```python
with open("books.json", "r") as file:
    data = json.load(file)
```

```python
for book in data["books"]:
    print("ID:", book["id"])
    print("Title:", book["title"])
    print("Author:", book["author"])
    print("Copies:", book["details"]["copies"])
    print("Available:", book["details"]["available"])
    print("\n")
```
